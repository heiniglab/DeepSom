```python
import pandas as pd
import re

df = pd.read_csv('/storage/groups/epigenereg01/workspace/projects/vale/datasets/snvs/BLCA-US/gnomAD_thr_0/vcfs/new_train.vcf', 
                 comment='#', sep='\t', names=['chrom', 'pos', 'id', 'ref', 'alt', 'qual', 'filter', 'info'])
```

```python
df['BAM'] = df['info'].apply(lambda x: re.search('BAM=([^;]*)',x).groups(1)[0] if 'BAM=' in x else None)
```

```python
df['GERMLINE'] = df['info'].apply(lambda x: True if 'GERMLINE' in x else False)
```

```python
df['true_label'] = df['info'].apply(lambda x: False if 'NON-SOMATIC' in x else True if 'SOMATIC' in x else None)
```

```python
df['vartype'] = df[['ref','alt']].apply(lambda x: 'SNP' if  len(x.ref)==len(x.alt) else 'INDEL', axis=1)
```

```python
df.sort_values(by='vartype', inplace=True)

df.true_label[1:20]
```

```python
pd.concat([df[df.vartype=='SNP'].sample(frac=1,random_state=1), df[df.vartype=='INDEL'].sample(frac=1,random_state=1)])
```
